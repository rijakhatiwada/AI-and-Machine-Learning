```python
def convert_units(conversion_type, value):
    """
    Converts units based on the conversion type provided.

    Parameters:
    conversion_type (str): The type of conversion to perform.
                           Options: m_to_ft, ft_to_m, kg_to_lbs, lbs_to_kg,
                                    l_to_gal, gal_to_l
    value (float): The numeric value to convert.

    Returns:
    float: The converted value.
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")


def main():
    print("Unit Conversion Program")
    print("-----------------------")
    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    try:
        choice = int(input("Enter your choice (1-3): "))

        if choice == 1:
            print("Length Conversion:")
            print("a. Meters to Feet")
            print("b. Feet to Meters")
            option = input("Choose option (a/b): ").lower()

            value = float(input("Enter value: "))

            if option == "a":
                result = convert_units("m_to_ft", value)
                print(f"{value} meters = {result:.2f} feet")
            elif option == "b":
                result = convert_units("ft_to_m", value)
                print(f"{value} feet = {result:.2f} meters")
            else:
                print("Invalid option.")

        elif choice == 2:
            print("Weight Conversion:")
            print("a. Kilograms to Pounds")
            print("b. Pounds to Kilograms")
            option = input("Choose option (a/b): ").lower()

            value = float(input("Enter value: "))

            if option == "a":
                result = convert_units("kg_to_lbs", value)
                print(f"{value} kg = {result:.2f} lbs")
            elif option == "b":
                result = convert_units("lbs_to_kg", value)
                print(f"{value} lbs = {result:.2f} kg")
            else:
                print("Invalid option.")

        elif choice == 3:
            print("Volume Conversion:")
            print("a. Liters to Gallons")
            print("b. Gallons to Liters")
            option = input("Choose option (a/b): ").lower()

            value = float(input("Enter value: "))

            if option == "a":
                result = convert_units("l_to_gal", value)
                print(f"{value} L = {result:.2f} gallons")
            elif option == "b":
                result = convert_units("gal_to_l", value)
                print(f"{value} gallons = {result:.2f} L")
            else:
                print("Invalid option.")

        else:
            print("Invalid choice.")

    except ValueError:
        print("Error: Please enter a valid numeric value.")


# Run the program
main()
```

```python
def calculate_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def calculate_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The maximum number.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The minimum number.
    """
    return min(numbers)


def main():
    print("Mathematical Operations Program")
    print("------------------------------")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = int(input("Enter your choice (1-4): "))

        # Input numbers separated by spaces
        user_input = input("Enter numbers separated by spaces: ")
        numbers = [float(num) for num in user_input.split()]

        if len(numbers) == 0:
            raise ValueError("List is empty.")

        if choice == 1:
            result = calculate_sum(numbers)
            print("Sum =", result)

        elif choice == 2:
            result = calculate_average(numbers)
            print("Average =", result)

        elif choice == 3:
            result = find_maximum(numbers)
            print("Maximum =", result)

        elif choice == 4:
            result = find_minimum(numbers)
            print("Minimum =", result)

        else:
            print("Invalid choice.")

    except ValueError:
        print("Error: Please enter valid numbers and ensure the list is not empty.")


# Run program
main()
```

```python
def extract_every_other(lst):
    """
    Extracts every other element from a list starting from the first element.

    Parameters:
    lst (list): The input list.

    Returns:
    list: A new list containing every other element.
    """
    return lst[::2]


# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = extract_every_other(numbers)
print("Output:", result)
```

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from the given list starting at index 'start'
    and ending at index 'end' (inclusive).

    Parameters:
    lst (list): The original list.
    start (int): Starting index.
    end (int): Ending index (inclusive).

    Returns:
    list: The extracted sublist.
    """
    return lst[start:end + 1]


# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = get_sublist(numbers, 2, 4)
print("Output:", result)
```

```python
def reverse_list(lst):
    """
    Reverses a list using slicing.

    Parameters:
    lst (list): The input list.

    Returns:
    list: The reversed list.
    """
    return lst[::-1]


# Example usage
numbers = [1, 2, 3, 4, 5]
result = reverse_list(numbers)
print("Output:", result)
```

```python
def remove_first_last(lst):
    """
    Removes the first and last elements from a list using slicing.

    Parameters:
    lst (list): The input list.

    Returns:
    list: A new list without the first and last elements.
    """
    return lst[1:-1]  # Slice from index 1 to second last element


# Example usage
numbers = [1, 2, 3, 4, 5]
result = remove_first_last(numbers)
print("Output:", result)
```

```python
def get_first_n(lst, n):
    """
    Extracts the first n elements from a list using slicing.

    Parameters:
    lst (list): The input list.
    n (int): Number of elements to extract.

    Returns:
    list: A new list containing the first n elements.
    """
    return lst[:n]  # Slice from start to n elements


# Example usage
numbers = [1, 2, 3, 4, 5]
result = get_first_n(numbers, 3)
print("Output:", result)
```

```python
def get_last_n(lst, n):
    """
    Extracts the last n elements from a list using slicing.

    Parameters:
    lst (list): The input list.
    n (int): Number of elements to extract from the end.

    Returns:
    list: A new list containing the last n elements.
    """
    return lst[-n:]  # Slice from the last n elements


# Example usage
numbers = [1, 2, 3, 4, 5]
result = get_last_n(numbers, 2)
print("Output:", result)
```

```python
def reverse_skip(lst):
    """
    Extracts every second element from a list starting from the second-to-last element,
    moving backward.

    Parameters:
    lst (list): The input list.

    Returns:
    list: A new list with elements in reverse order, skipping one in between.
    """
    return lst[-2::-2]  # Start from second-to-last, step -2 backwards


# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = reverse_skip(numbers)
print("Output:", result)
```

```python
def flatten(lst):
    """
    Flattens a nested list into a single list with all elements in one dimension.

    Parameters:
    lst (list): A nested list (list of lists).

    Returns:
    list: A flattened list containing all elements.
    """
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list


# Example usage
nested_list = [[1, 2], [3, 4], [5]]
result = flatten(nested_list)
print("Output:", result)
```

```python
def access_nested_element(lst, indices):
    """
    Accesses a specific element from a nested list given a list of indices.

    Parameters:
    lst (list): The nested list.
    indices (list): A list of indices specifying the position of the element.

    Returns:
    The element at the specified position.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element


# Example usage
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
indices = [1, 2]
result = access_nested_element(nested_list, indices)
print("Output:", result)
```

```python
def sum_nested(lst):
    """
    Recursively calculates the sum of all numbers in a nested list, regardless of depth.

    Parameters:
    lst (list): The nested list.

    Returns:
    int/float: Sum of all elements in the nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)  # Recursively sum nested lists
        else:
            total += item
    return total


# Example usage
nested_list = [[1, 2], [3, [4, 5]], 6]
result = sum_nested(nested_list)
print("Output:", result)
```

```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): The nested list.
    elem: The element to remove.

    Returns:
    list: A new nested list with the element removed.
    """
    new_list = []
    for sublist in lst:
        if isinstance(sublist, list):
            filtered_sublist = [x for x in sublist if x != elem]
            new_list.append(filtered_sublist)
        else:
            # Handle case where lst itself has non-list elements
            if sublist != elem:
                new_list.append(sublist)
    return new_list


# Example usage
nested_list = [[1, 2], [3, 2], [4, 5]]
element_to_remove = 2
result = remove_element(nested_list, element_to_remove)
print("Output:", result)
```

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list regardless of depth.

    Parameters:
    lst (list): The nested list.

    Returns:
    int/float: The maximum element in the nested list.
    """
    maximum = float('-inf')  # Start with smallest possible value
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))  # Recursively find max
        else:
            maximum = max(maximum, item)
    return maximum


# Example usage
nested_list = [[1, 2], [3, [4, 5]], 6]
result = find_max(nested_list)
print("Output:", result)
```

```python
def count_occurrences(lst, elem):
    """
    Counts how many times a specific element appears in a nested list.

    Parameters:
    lst (list): The nested list.
    elem: The element to count.

    Returns:
    int: The number of occurrences of the element in the nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)  # Recursively count in sublists
        elif item == elem:
            count += 1
    return count


# Example usage
nested_list = [[1, 2], [2, 3], [2, 4]]
element_to_count = 2
result = count_occurrences(nested_list, element_to_count)
print("Output:", result)
```

```python
def deep_flatten(lst):
    """
    Recursively flattens a deeply nested list into a single list, regardless of depth.

    Parameters:
    lst (list): The deeply nested list.

    Returns:
    list: A flattened list containing all elements.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))  # Recursively flatten sublists
        else:
            flat_list.append(item)
    return flat_list


# Example usage
nested_list = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
result = deep_flatten(nested_list)
print("Output:", result)
```

```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
    lst (list): The nested list.

    Returns:
    float: The average of all elements.
    """
    def flatten(lst):
        """Helper function to flatten the nested list completely."""
        flat_list = []
        for item in lst:
            if isinstance(item, list):
                flat_list.extend(flatten(item))
            else:
                flat_list.append(item)
        return flat_list

    flat_list = flatten(lst)
    if not flat_list:
        return 0  # Avoid division by zero
    return sum(flat_list) / len(flat_list)


# Example usage
nested_list = [[1, 2], [3, 4], [5, 6]]
result = average_nested(nested_list)
print("Output:", result)
```
